```python
#CELDA 1 - MENSAJES BIENVENIDA
# # 🚀 Misión Galáctica
# Bienvenida
print("Bienvenid@s a la misión galáctica 🪐\n")

from IPython.display import HTML, display, clear_output
import time
import random

# Mensaje inicial con colores
display(HTML("<div style='padding:10px;border-radius:10px;background:#1f2937;color:#fff'> 🚀 Vamos a iniciar el entrenamiento de la tripulación...</div>"))
```

```python

# CELDA 2 - VARIABLES Y VERIFICACIÓN
# Variables de la capitana
capitana = 'Arantxa'
energia = 75
escudo_activo = True
print(f"{capitana} {energia} {escudo_activo}\n")

# Verificar tipos
assert isinstance(capitana, str), 'capitana debe ser texto'
assert isinstance(energia, (int, float)), 'energia debe ser número'
assert isinstance(escudo_activo, bool), 'escudo_activo debe ser booleano'
print('✅ Reto 1 OK\n')

# Mensaje estado capitana
estado_escudo = "activo" if escudo_activo else "inactivo"
mensaje = f"La capitana {capitana} tiene {energia} de energía y su escudo está {estado_escudo}."
print(mensaje)
print('✅ Reto 1 OK\n')

```

```python
#CELDA 3 - CONDICIONALES - CONDICIONAL DE ATERRIZAJE
# Condicional aterrizaje
energia = 45
escudo = 60
if energia >= 40 and escudo >= 50:
    print("🛬 Aterrizaje autorizado ✅\n")
else:
    print("⛔ Aterrizaje denegado\n")
```

```python

#CELDA 4 - BUCLES - REVISAR MÓDULOS
# Recorrer módulos
modulos = ['motor', 'radar', 'escudos', 'comunicaciones', "panel", "apertura automática"]
for modulo in modulos:
    print(f"🔍 Revisando módulo: {modulo}")
print()
```

```python

#CELDA 5 - FUNCIONES - PUNTOS DE MISIÓN
# Función puntos_mision
def puntos_mision(tareas, bonus):
    return tareas * 10 + bonus

print('Puntos:', puntos_mision(6, 12))
assert puntos_mision(2, 5) == 25
print('✅ Reto 4 OK\n')
```

```python

#CELDA 6 - DICCIONARIO - DICCIONARIO PILOTA
# Diccionario piloto
pilota = {
    "nombre": "Arantxa",
    "rol": "Capitana",
    "nivel": 5
}
print(f"{pilota['nombre']} tiene el rol de {pilota['rol']}\n")
```

```python

#CELDA 7 - CREATIVO - BARRA PROGRESO
#Barra de progreso animada
progreso_final = 75  # Valor final de la barra
for progreso in range(progreso_final + 1):
    clear_output(wait=True)
    display(HTML(f"""
    <div style='font-family:system-ui;padding:5px'>
        <p>Progreso: <b>{progreso}%</b></p>
        <div style='width:320px;height:18px;border-radius:10px;background:#ddd'>
            <div style='width:{progreso}%;height:18px;border-radius:10px;background:#22c55e'></div>
        </div>
    </div>
    """))
    time.sleep(0.05)
```

```python


#CELDA 8 - RANDOM - ELEGIR PLANETA AL AZAR
# Elegir planeta al azar
planetas = ['🪐 Saturno', '🔴 Marte', '🌕 Luna', '🌌 Nebulosa X']
destino = random.choice(planetas)
print(f"🚀 Preparando la nave... ¡Nuestro destino será {destino}!\n")

```

```python
#CELDA 9 - EXTRA FOR IN - CUENTA ATRÁS Y DESPEGUE
# Cuenta atrás antes del despegue
print("🚀 Preparando el despegue...")
for i in range(5, 0, -1):
    print(f"{i}...")
    time.sleep(1)
print("¡Despegue! 🚀")
```
