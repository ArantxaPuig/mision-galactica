# 🪐 ¿Qué es Jupyter Notebook?

**Jupyter Notebook** es un **entorno interactivo** donde podemos escribir y ejecutar **código Python**.

Se divide en **celdas**, que puedes ejecutar **una por una**.

Permite mezclar en el mismo documento:

- 💻 Código
- 📊 Resultados
- 📝 Explicaciones en texto

Esto lo hace **perfecto para aprender, experimentar y tomar apuntes**.

---

## 📄 Extensión de los archivos

Los archivos de Jupyter tienen la extensión:

.ipynb


Significa **IPython Notebook**.

---

# ⭐ Ventajas de Jupyter Notebook

- Ver el **resultado de cada celda inmediatamente**
- Escribir **explicaciones y anotaciones**
- Mantener el trabajo **organizado**
- Crear documentos **reproducibles**

# 🧠 Estructuras básicas de Python usadas en el ejercicio

En el ejercicio **“Misión Galáctica”** utilizamos varias estructuras fundamentales de Python.

---

# 1️⃣ Variables

Las **variables** sirven para **guardar información**.

Tipos comunes:

- números
- texto
- valores booleanos (True / False)

Ejemplo:

energia = 75
capitana = "Arantxa"
escudo_activo = True


| Variable | Tipo |
|------|------|
| energia | Número |
| capitana | Texto |
| escudo_activo | Booleano |

---

# 2️⃣ Print y f-strings

La función `print()` se usa para **mostrar información en pantalla**.

Las **f-strings** permiten combinar texto con variables.

Ejemplo:

print(f"{capitana} tiene {energia} de energía")


Resultado:

Arantxa tiene 75 de energía


---

# 3️⃣ Condicionales (if / else)

Los **condicionales** permiten ejecutar código dependiendo de una condición.

Ejemplo:

if energia >= 40:
print("Aterrizaje autorizado")
else:
print("Aterrizaje denegado")


Funcionamiento:

| Condición | Resultado |
|------|------|
| energía suficiente | aterrizaje autorizado |
| energía baja | aterrizaje denegado |

---

# 4️⃣ Bucles (for)

Los **bucles** permiten repetir código varias veces.

Ejemplo: recorrer una lista de módulos.

modulos = ["motor", "radar", "escudos"]

for modulo in modulos:
print("Revisando", modulo)


Resultado:

Revisando motor
Revisando radar
Revisando escudos


---

# 5️⃣ Funciones

Las **funciones** son bloques de código reutilizables que realizan una tarea.

Ejemplo:

def puntos_mision(tareas, bonus):
return tareas * 10 + bonus


Uso de la función:

print(puntos_mision(6,12))


Resultado:

72


---

# 6️⃣ Diccionarios

Los **diccionarios** almacenan información en pares **clave-valor**.

Ejemplo:

pilota = {
"nombre": "Arantxa",
"rol": "Capitana"
}


Acceder a un valor:

print(pilota["nombre"])


Resultado:

Arantxa


---

# 7️⃣ Listas y selección aleatoria

Las **listas** permiten guardar varios elementos.

Ejemplo:

planetas = ["Marte","Luna","Saturno"]


Podemos seleccionar uno **al azar** usando `random`.

import random

destino = random.choice(planetas)


Esto elegirá un planeta aleatorio.

---

# 8️⃣ Animaciones y temporizadores

En Jupyter podemos hacer **animaciones simples**.

Para ello usamos:

time.sleep()


Esto crea **pausas entre instrucciones**.

También podemos usar:

clear_output()


Para **limpiar la salida de la celda y simular animaciones**.
