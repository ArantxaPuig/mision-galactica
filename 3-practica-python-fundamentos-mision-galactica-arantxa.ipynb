## Reto 1 — Variables y tipos
Crea estas variables: `capitana` (texto), `energia` (número), `escudo_activo` (booleano).

```python
# variables
capitana = 'Arantxa'
energia = 75
escudo_activo = True
print(capitana, energia, escudo_activo)
```

```python
assert isinstance(capitana, str), 'capitana debe ser texto'
assert isinstance(energia, (int, float)), 'energia debe ser número'
assert isinstance(escudo_activo, bool), 'escudo_activo debe ser booleano'
print('✅ Reto 1 OK')
```

```python
# Paso 1: Crear variables
capitana = 'Arantxa'
energia = 75
escudo_activo = True

# Paso 2: Verificar tipos
assert isinstance(capitana, str), 'capitana debe ser texto'
assert isinstance(energia, (int, float)), 'energia debe ser número'
assert isinstance(escudo_activo, bool), 'escudo_activo debe ser booleano'

# Paso 3: Crear mensaje
estado_escudo = "activo" if escudo_activo else "inactivo"
mensaje = f"La capitana {capitana} tiene {energia} de energía y su escudo está {estado_escudo}."

# Paso 4: Mostrar mensaje
print(mensaje)
print('✅ Reto 1 OK')
```

## Reto 2 — Condicionales
Muestra `Aterrizaje autorizado` si `energia >= 40` y `escudo >= 50`; si no, denegado.

## Reto 4 — Funciones
Completa `puntos_mision` para que devuelva `tareas*10 + bonus`.

```python
def puntos_mision(tareas, bonus):
    puntos = tareas * 10 + bonus
    return puntos

print('Puntos:', puntos_mision(6, 12))
```

```python
assert puntos_mision(2, 5) == 25, 'Revisa la fórmula de la función'

print('✅ Reto 4 OK')
```

## Reto 6 (creativo) — Barra de progreso
Asigna un valor a `progreso` (0-100) y visualiza una barra.

```python
from IPython.display import display, HTML, clear_output
import time

# Valor final de progreso
progreso_final = 75  # Cambia entre 0 y 100

# Animar la barra
for progreso in range(progreso_final + 1):
    clear_output(wait=True)
    display(HTML(f"""
    <div style='font-family:system-ui'>
        <p>Progreso: <b>{progreso}%</b></p>
        <div style='width:320px;height:18px;border-radius:10px;background:#ddd'>
            <div style='width:{progreso}%;height:18px;border-radius:10px;background:#22c55e'></div>
        </div>
    </div>
    """))
    time.sleep(0.05)
```
